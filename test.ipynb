```python
from a3c.envs import create_unreal_env
from pspnet.utils import color_class_image
from dragonfly.DragonflyGym import DragonflyEnv
import matplotlib.pyplot as plt
import numpy as np
from IPython import display
import time
```

```python
%matplotlib inline
```

```python
env1 = create_unreal_env()
```

```python
obs = env1.reset()
time.sleep(1)
done = False
img = plt.imshow(color_class_image(np.argmax(obs.transpose(1, 2, 0), axis=2)))
plt.axis('off')
display.display(plt.gcf())
display.clear_output(wait=True)

while not done:
    obs, reward, done, info = env1.step(0)
    img.set_data(color_class_image(np.argmax(obs.transpose(1, 2, 0), axis=2)))
    plt.axis('off')
    display.display(plt.gcf())
    display.clear_output(wait=True)
```
